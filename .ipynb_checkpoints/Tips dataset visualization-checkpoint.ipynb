```python
# Packages / libraries
import os
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Loading dataset
tips = sns.load_dataset("tips")
tips
```

```python
sns.set(style="whitegrid", palette="pastel", color_codes=True)
sns.set(rc={'figure.figsize':(12,8)})
```

```python
#Violinplot 
sns.violinplot(x="day", y="total_bill", hue="smoker",
               split=True, inner="quartile",
               palette={"Yes": "y", "No": "b"},
               data=tips)
sns.despine(left=True)
```

```python
#Heatmap 

pc = tips[['total_bill','tip',"size"]].corr(method = "pearson")

cols = ['total_bill CC',"tip CC", "size CC"]

ax = sns.heatmap(pc, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 50},
                 cmap="Blues")
```

```python
# JoinPlots- Tip vs Totall bill relationship

g = sns.jointplot("tip", "total_bill", data=tips, kind="reg", color = 'red', height = 10)
```

```python
# Pairplot
g = sns.pairplot(tips[['total_bill','tip','size','sex']], hue = 'sex', kind="reg",height = 6)
```

```python
# Pairplot running on Totall bill, Tip and Size column adding hue-Smoker 
g = sns.pairplot(tips[['total_bill','tip','size','smoker']], hue = 'smoker')
```

```python
# Pairplot running on Total bill, Tip, Size column
g = sns.pairplot(tips[['total_bill','tip','size']], plot_kws={'color':'#0EDCA8'})
```

```python
# Pairplot running on all detaset, color red
g = sns.pairplot(tips, plot_kws={'color':'red'})
```

```python
# SubPlots

fig, axes = plt.subplots(2, 2, figsize=(12, 7))

a = tips['total_bill'].values
b = tips['tip'].values
c = tips['size'].values


# plot 1

sns.distplot(a, color = 'blue', ax=axes[0,0])
axes[0,0].set_title = 'total_bill'

# plot 2
sns.distplot(b, color = 'blue', ax=axes[0,1])

# plot 3
sns.distplot(c, color = 'blue', ax=axes[1,0])

# plot 4
sns.boxplot(x="total_bill", y="day", data=tips, color = '#52F954', ax=axes[1,1])
sns.swarmplot(x="total_bill", y="day", data=tips, color="blue", ax=axes[1,1])
```

```python
# ImPlot- relatonship between Tip and Total bill
ax = sns.lmplot(x="total_bill", y="tip", data=tips, height=9)
```

```python
# Implot Relationship between Total bill and tip - changing color, hue & Style
ax = sns.lmplot(x="total_bill", y="tip", data=tips, hue = 'sex', ci= False, height=7, markers=["o", "x"])
```

```python
# ImPlot Relationship between Totall bil and tip - by column Sex
ax = sns.lmplot(x="total_bill", y="tip", data=tips, col = 'sex', ci= False, height=5, 
                line_kws={'color': 'red'}, 
                scatter_kws={'color':'#ADC11E'})
```

```python
# Implot - Relationship between Size and Tip - by column - by day - add Jitter too
ax = sns.lmplot(x="size", y="tip", data=tips, col = 'day', ci= False, height=4, 
                line_kws={'color': '#031722'}, 
                scatter_kws={'color':'#1E84C1'},
               col_wrap=3,
               x_jitter=.3)
```

```python
# ScatterPlot - Relationship bteween Total bill and Tip including hue Smoker and size variable for Sex 
ax = sns.scatterplot(x="total_bill", y="tip", data=tips, color = 'green', hue = 'smoker', size = 'sex',
                    sizes=(20, 200))
```

```python
# # ScatterPlot - Relationship between Total bill and tip- including hue and style
ax = sns.scatterplot(x="total_bill", y="tip", data=tips, hue = 'sex', style = 'sex')
```

```python
## ScatterPlot - Relationship bteween Total bill and Tip
ax = sns.scatterplot(x="total_bill", y="tip", data=tips, color = 'green')
```

```python
#BoxPlot- Investigating the distribution of Total bill

x = tips['total_bill'].values

ax = sns.boxplot(x)

print('The meadian is: ', tips['total_bill'].median())
```

```python
# BoxPlot- Investigating the distribution of Tip by Sex
ax = sns.boxplot(x="sex", y="tip", data=tips, color = '#D1EC46')
ax = sns.swarmplot(x="sex", y="tip", data=tips, color="red")
```

```python
# BoxPlot - Investigating the distribution of Tip by Day adding hue-Sex
ax = sns.boxplot(x="day", y="tip", data=tips, color="#B971E7", hue = 'sex')
ax.legend(loc='upper center')
```

```python
# BoxPlot - Investigating the distribution of Tip by Day - Horizontal

ax = sns.boxplot(x="tip", y="day", data=tips, color = '#14B29A')
```

```python
#BoxPlot - Investigating the distribution of Total bil by Day
ax = sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# Histogram-  Investigating the distribution of Total bill

x = tips['total_bill'].values

sns.distplot(x, color = 'orange');
```

```python
# Histogram-  Investigating the distribution of Total bill adding the mean

x = tips['total_bill'].values

sns.distplot(x, color = 'orange');

mean = tips['total_bill'].mean()

plt.axvline(mean, 0,1, color = 'red')
```

```python
# Histogram-  Investigating the distribution of Tip

x = tips['tip'].values

sns.distplot(x, color = 'red');

mean = tips['tip'].mean()

plt.axvline(mean, 0,1, color = 'blue')
```

```python
# BarPlot - Total Tip by Day

ax = sns.barplot(x="day", y="tip", data=tips)

tips[['day', 'tip']].groupby('day', as_index = False).agg({'tip':'mean'})
```

```python
# BarPlot - Total Total bill by Day

ax = sns.barplot(x="day", y="total_bill", data=tips, ci=False)
```

```python
# BarPlot - Total Total bill by Day, adding hue Sex

ax = sns.barplot(x="day", y="total_bill", data=tips, ci=False, hue = 'sex')
```

```python
# BarPlot - Total Total bill by Day, Horizontal

ax = sns.barplot(x="total_bill", y="size", ci=False, hue = 'sex', orient = 'h', data=tips, color="#71E418")
```
